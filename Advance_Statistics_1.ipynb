```python
#1.What is a random variable in probability theory?
'''
Random Variable in Probability Theory:

Definition: 
   A random variable assigns a numerical value to each outcome of a random experiment.

Purpose:
   It allows random outcomes to be studied using mathematical methods.

Relation to Sample Space:
   Each outcome in the sample space is mapped to a real number.
'''
```

```python
#2 What are the types of random variables?
'''
Types od Random Variables :

Discrete: Takes countable values (e.g., number of heads in three coin tosses).
Continuous: Takes values over an interval (e.g., time to complete an exam).

Notation
Random variables are denoted by capital letters (e.g., (X)); their values by lowercase letters 
   (e.g., (x)).

Example:
   Tossing a die once: let (X) be the number obtained. Then (X \in {1,2,3,4,5,6}), each with probability 
   (1/6).
'''
```

```python
#3. Explain the difference between discrete and continuous distributions.
'''
Difference Between Discrete and Continuous Distributions

Nature of Values:

Discrete distribution : The random variable takes a finite or countable number of distinct values.
Continuous distribution : The random variable can take any value within a continuous range.

Probability Assignment:
Discrete: Probabilities are assigned to individual values.
Continuous: Probabilities are assigned to intervals, not to exact values.

Probability at a Point:
Discrete: (P(X = x) > 0) is possible.
Continuous: (P(X = x) = 0) for any specific value.

Mathematical Representation:

Discrete: Described using a probability mass function (PMF).
Continuous: Described using a probability density function (PDF).

Examples

Discrete: Number of defective items in a batch.
Continuous: Height or weight of a person.

Graphical Form
Discrete: Shown using bar graphs.
Continuous: Shown using smooth curves.
'''
```

```python
#4.What is a binomial distribution, and how is it used in probability?
'''
Binomial Distribution

1. Definition:
   A binomial distribution models the number of successes in a fixed number of independent trials, where each trial has 
   two possible outcomes (success or failure) and a constant probability of success.

2. Conditions:

* Fixed number of trials (n)
* Only two outcomes per trial
* Trials are independent
* Probability of success (p) is the same for each trial

3. Use in Probability:
   It is used to calculate the probability of obtaining a specific number of successes in repeated experiments.

4. Example:
   Toss a fair coin 5 times. Let success be getting a head. The binomial distribution gives the probability of getting exactly 
   3 heads out of 5 tosses.

5. Importance:
   It is widely applied in quality control, surveys, and decision-making problems involving yes/no outcomes.
   
'''
```

```python
#5.What is the standard normal distribution, and why is it important?
'''
Standard Normal Distribution

1. Definition
   The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1.

2. Characteristics
   It is symmetric and bell-shaped, with most values concentrated around the mean.

3. Standardization
   Any normal random variable can be converted to the standard normal form using a z-score.

4. Use in Probability
   It is used to find probabilities and critical values for normally distributed data.

5. Importance
   The standard normal distribution allows comparison across different normal distributions and is fundamental in statistical 
   inference and hypothesis testing.
'''
```

```python
#6.What is the Central Limit Theorem (CLT), and why is it critical in statistics?
'''
Central Limit Theorem (CLT)

1. Definition
   The Central Limit Theorem states that the distribution of the sample mean approaches a normal distribution as the sample size 
   increases, regardless of the populationâ€™s original distribution.

2. Conditions
   It applies when samples are random and sufficiently large, with independent observations.

3. Key Result
   For large samples, the mean of the sampling distribution equals the population mean, and its standard deviation equals the population 
   standard deviation divided by the square root of the sample size.

4. Use in Statistics
   The CLT allows normal probability methods to be used for estimating parameters and testing hypotheses.

5. Importance
   It is critical because it justifies many statistical techniques and enables reliable inference from sample data.

'''
```

```python
#7.What is the significance of confidence intervals in statistical analysis?
'''
Significance of Confidence Intervals in Statistical Analysis

1. Definition:
   A confidence interval is a range of values, calculated from sample data, that is likely to contain the true population parameter.

2. Measure of Uncertainty:
   It shows the uncertainty associated with a sample estimate rather than providing a single point estimate.

3. Interpretation:
   A stated confidence level (such as 95%) indicates how reliable the interval is in capturing the true parameter over repeated samples.

4. Use in Inference:
   Confidence intervals are used to estimate population parameters and assess the precision of estimates.

5. Importance:
   They support informed decision-making by indicating both the estimated value and its possible variation.
'''
```

```python
#8.Concept of Expected Value in a Probability Distribution
'''
1. Definition
   The expected value is the weighted average of all possible values of a random variable, where the weights are their respective 
   probabilities.

2. Meaning
   It represents the long-run average outcome if the random experiment is repeated many times.

3. Calculation
   For a discrete distribution, it is calculated by summing the product of each value and its probability.

4. Example
   If a fair die is rolled, the expected value is
   (1+2+3+4+5+6)/6 = 3.5.

5. Importance
   Expected value is used to compare outcomes, evaluate risk, and support decision-making under uncertainty.
'''
```

```python
#9.Python code for 1000 random variables , mean, std dev, and histogram.


import numpy as np                  # Import NumPy for numerical operations
import matplotlib.pyplot as plt     # Import Matplotlib for plotting

data = np.random.normal(50, 5, 1000)   # Generate 1000 random values (mean=50, std=5)

mean = np.mean(data)                # Calculate the mean of the data
std_dev = np.std(data)              # Calculate the standard deviation of the data

print("Mean:", mean)                # Display the mean
print("Standard Deviation:", std_dev)  # Display the standard deviation

plt.hist(data, bins=30)             # Create histogram with 30 bins
plt.xlabel("Value")                 # Label x-axis
plt.ylabel("Frequency")             # Label y-axis
plt.title("Histogram of Normal Distribution")  # Add title
plt.show()                          # Display the plot
```

```python
#10.Python Code to Compute Mean and 95% Confidence Interval

import numpy as np                                  # Import NumPy for calculations

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

mean_sales = np.mean(daily_sales)                   # Calculate sample mean
std_sales = np.std(daily_sales, ddof=1)             # Calculate sample standard deviation
n = len(daily_sales)                                # Sample size

standard_error = std_sales / np.sqrt(n)             # Compute standard error
z = 1.96                                            # Z-value for 95% confidence level

lower_ci = mean_sales - z * standard_error          # Lower confidence limit
upper_ci = mean_sales + z * standard_error          # Upper confidence limit

print("Mean Daily Sales:", mean_sales)
print("95% Confidence Interval:", (lower_ci, upper_ci))
```
