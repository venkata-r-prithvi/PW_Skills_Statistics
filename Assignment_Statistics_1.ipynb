```python
'''
1) Descriptive statistics vs. Inferential statistics

Descriptive statistics:
Descriptive statistics are used to summarize and present the main features of a dataset. They do not go 
beyond the data that is actually observed. The goal is to make large or complex data easier to understand.

Common tools include:

Measures of central tendency: mean, median, mode

Measures of dispersion: range, variance, standard deviation

Graphical methods: bar charts, histograms, box plots

Example:
If a class of 60 students scores marks in an exam, descriptive statistics can be used to calculate the 
average mark, identify the highest and lowest scores, and show the distribution of marks using a 
histogram. These results only describe the performance of that particular class.

Inferential statistics

Inferential statistics are used to draw conclusions or make predictions about a population based on a 
sample. They involve probability and statistical tests to account for uncertainty.

Common tools include:

Hypothesis testing (t-test, chi-square test, ANOVA)

Confidence intervals

Regression analysis

Example:

If marks from a sample of 60 students are taken from a university with 2,000 students, inferential 
statistics can be used to estimate the average mark of all students or to test whether a new teaching 
method has significantly improved performance.

'''
```

```python
'''
2) Sampling in Statistics

Sampling is the process of selecting a subset of a population to analyze and draw conclusions about 
the entire population. It is used when studying the whole population is impractical due to time or cost 
constraints.

Random Sampling

Random sampling gives every population member an equal chance of selection.

-Selection is purely by chance

-Minimizes selection bias

-Best for relatively homogeneous populations

Example:
Selecting 100 employees at random from a company of 1,000.

Stratified Sampling

Stratified sampling divides the population into subgroups (strata) and randomly samples from each group, 
usually in proportion to size.

-Ensures representation of all key subgroups

-More accurate for heterogeneous populations

Example:
If a college has 60% undergraduates and 40% postgraduates, a stratified sample of 100 includes 60 
undergraduates and 40 postgraduates.
'''
```

```python
'''
3) Mean, Median, and Mode

Mean
The mean is the arithmetic average of a dataset, calculated by dividing the sum of all values by the
number of observations.
Example: For values 10, 20, and 30, the mean is (10 + 20 + 30) / 3 = 20.

Median
The median is the middle value when data is arranged in ascending or descending order. If there is an 
even number of observations, it is the average of the two middle values.
Example: For 5, 10, 15, the median is 10.

Mode
The mode is the most frequently occurring value in a dataset. A dataset may have one mode, more than 
one mode, or no mode.
Example: In 2, 4, 4, 6, the mode is 4.
Importance of Measures of Central Tendency

These measures summarize data using a single representative value, making large datasets easier to
understand. The mean is useful for balanced data, the median is best when data contains outliers, and 
the mode helps identify the most common value. 

Together, they provide a clear picture of the data’s central behavior.

'''
```

```python
'''
4) Skewness:
Skewness measures the asymmetry of a data distribution around its mean. It indicates whether the data is 
skewed to the left or right.

Positive skew (right skew): The right tail is longer. Most values are concentrated on the lower side, 
with a few large values pulling the mean to the right.

Negative skew (left skew): The left tail is longer. Most values are concentrated on the higher side.

Kurtosis:
Kurtosis measures the peakedness and tail heaviness of a distribution compared to a normal distribution.

High kurtosis: Sharp peak and heavy tails (more extreme values).

Low kurtosis: Flat peak and light tails (fewer extreme values).

Implication of Positive Skew :

A positive skew implies that most observations are smaller, with a few large outliers, 
and typically mean > median > mode.

'''
```

```python
#5) Calculate Mean Median Mode 

from statistics import mean, median, mode

# Given data
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


mean_value = mean(numbers) # Calculations
median_value = median(numbers)
mode_value = mode(numbers)


print("Mean:", mean_value) # Mean , Median Mode Values 
print("Median:", median_value)
print("Mode:", mode_value)
```

```python
#6) Covariance and Correlation calculations 

import numpy as np


list_x = [10, 20, 30, 40, 50] # Given data
list_y = [15, 25, 35, 45, 60]


x = np.array(list_x) # Convert to numpy arrays
y = np.array(list_y)

# Compute covariance
cov_xy = np.cov(x, y, ddof=1)[0, 1]  # ddof=1 for sample covariance


corr_xy = np.corrcoef(x, y)[0, 1] # Compute correlation coefficient

# Output
print("Covariance:", cov_xy)
print("Correlation Coefficient:", corr_xy)

```

```python
'''
Covariance measures how two variables move together. Positive value → they increase together.

Correlation coefficient standardizes covariance to a range of -1 to 1, showing the strength and 
direction of linear relationship.

Here, 0.9959 indicates a very strong positive linear relationship.
'''
```

```python
#7) Box Plot 
import matplotlib.pyplot as plt
import numpy as np


data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35] # Given data


plt.boxplot(data, vert=False, patch_artist=True, showmeans=True) # Create a boxplot
plt.title("Boxplot of Data")
plt.xlabel("Values")
plt.show()


Q1 = np.percentile(data, 25) # Identify outliers using IQR method
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)

```

```python
'''
 boxplot visually shows the median, quartiles, and range of the data.

Q1 (25th percentile): 17, Q3 (75th percentile): 24.5

IQR (Interquartile Range): 7.5

Outlier bounds: Anything below 5.75 or above 35.75 is considered an outlier.

In this dataset, 35 is the largest value but not an outlier, as it falls within the upper bound.

Boxplots are helpful to quickly identify spread, skewness, and extreme values in data.

'''
```

```python
#8 Covariance and Corelation
import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute covariance
covariance = np.cov(advertising_spend, daily_sales, ddof=1)[0, 1]

# Compute correlation coefficient
correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
'''
Covariance (41250) → positive, meaning advertising spend and sales increase together.

Correlation (0.993) → very strong positive linear relationship.

Marketing can be confident that higher ad spend is strongly associated with higher daily sales.
'''
```
