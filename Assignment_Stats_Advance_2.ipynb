```python
#1.What is hypothesis testing in statistics?
'''
Hypothesis testing in statistics is a systematic method used to make decisions about a population based on sample data. It helps determine 
whether observed results are due to random variation or represent a meaningful effect.

1. Definition
   Hypothesis testing is a procedure for evaluating assumptions about a population parameter using sample data and probability theory.

2. Purpose
   The objective is to assess the validity of a claim and decide whether there is enough evidence to support it.

3. Statistical Hypotheses
   Two hypotheses are involved.
   The null hypothesis (H₀) assumes no effect or no difference.
   The alternative hypothesis (H₁) represents the presence of an effect or difference.

4. Test Statistic
   A test statistic is computed from the sample data to measure how much the observed outcome deviates from what is expected under the 
   null hypothesis.

5. Significance Level
   The significance level (α), commonly set at 0.05, defines the risk of rejecting a true null hypothesis.

6. P-value
   The p-value measures the probability of observing the sample results, or more extreme results, assuming the null hypothesis is true.

7. Decision Rule
   If the p-value is less than or equal to the significance level, the null hypothesis is rejected; otherwise, it is not rejected.

8. Conclusion
   A conclusion is drawn regarding whether sufficient statistical evidence exists to support the alternative hypothesis.


'''
```

```python
#2 What is the null hypothesis, and how does it differ from the alternative hypothesis?
'''
The null hypothesis and the alternative hypothesis are two complementary statements used in hypothesis testing to evaluate claims about 
a population.

1. Null Hypothesis
   The null hypothesis, denoted as H₀, states that there is no effect, no difference, or no relationship between variables. 
   It represents the default or status-quo assumption and is the hypothesis that is tested directly using statistical methods.

2. Alternative Hypothesis
   The alternative hypothesis, denoted as H₁ or Ha, states that there is an effect, a difference, or a relationship. 
   It represents the claim that contradicts the null hypothesis and is supported only if sufficient statistical evidence is found.

3. Key Differences
   The null hypothesis assumes no change or no impact, while the alternative hypothesis suggests a meaningful change or impact.
   The null hypothesis is rejected or not rejected, whereas the alternative hypothesis is accepted only indirectly when the null hypothesis
   is rejected.
   The null hypothesis includes an equality statement, while the alternative hypothesis includes inequality (greater than, less than, 
   or not equal to).

4. Purpose in Testing
   Together, these hypotheses provide a clear framework for decision-making based on sample data and probability, allowing researchers 
   to draw objective conclusions.
'''
```

```python
#3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
'''
The significance level is a key concept in hypothesis testing that defines the criteria for making decisions about statistical evidence.

1. Definition
   The significance level, denoted by α (alpha), is the maximum probability of rejecting the null hypothesis when it is actually true. 
   It represents the acceptable level of risk for making a Type I error.

2. Common Values
   In practice, common significance levels are 0.05, 0.01, and 0.10, depending on the field of study and the consequences of making an 
   incorrect decision.

3. Role in Hypothesis Testing
   The significance level serves as a benchmark for evaluating the strength of evidence against the null hypothesis. It is set before
   analyzing the data to ensure objectivity in decision-making.

4. Relationship with the P-value
   The p-value is compared with the significance level to determine the test outcome. If the p-value is less than or equal to α,
   the null hypothesis is rejected. If the p-value is greater than α, the null hypothesis is not rejected.

5. Impact on Decision Making
   A lower significance level requires stronger evidence to reject the null hypothesis, reducing the chance of false positives. 
   A higher significance level makes rejection easier but increases the risk of error.

6. Conclusion
   The significance level plays a critical role in controlling error and guiding consistent, data-driven conclusions in hypothesis testing.
'''
```

```python
#4.What are Type I and Type II errors? Give examples of each.
'''
Type I and Type II errors are possible mistakes that can occur during hypothesis testing when making decisions based on sample data.

1. Type I Error
   A Type I error occurs when the null hypothesis is rejected even though it is actually true. In other words, it involves concluding 
   that an effect or difference exists when it does not.

Example
A company concludes that a new marketing strategy increases sales when, in reality, it has no real effect. This is a false positive result.

2. Type II Error
   A Type II error occurs when the null hypothesis is not rejected even though it is false. This means failing to detect an effect or
   difference that actually exists.

Example
A company concludes that a new marketing strategy does not affect sales when, in fact, it does lead to increased sales. This is a 
false negative result.

3. Key Difference
   Type I error is related to rejecting a true null hypothesis, while Type II error is related to not rejecting a false null hypothesis.

4. Importance
   Understanding these errors helps researchers choose appropriate significance levels and sample sizes to balance the risks of 
   incorrect decisions.
   
'''
```

```python
#5. What is the difference between a Z-test and a T-test? Explain when to use each.

'''
The Z-test and T-test are statistical methods used in hypothesis testing to compare sample data with population parameters or
between groups. They differ mainly in the assumptions they make about the population and sample size.

1. Z-test
   A Z-test is used when the population variance is known and the sample size is large, typically 30 or more. It assumes that the sampling 
   distribution of the mean follows a normal distribution or that the population itself is normally distributed. Z-tests are commonly used 
   for testing population means and proportions.

When to use
A Z-test is appropriate when the population standard deviation is known and the sample size is sufficiently large.

2. T-test
   A T-test is used when the population variance is unknown and must be estimated from the sample data. It is suitable for small sample 
   sizes, generally less than 30, and assumes that the population is approximately normally distributed. The T-test accounts for additional
   uncertainty by using the t-distribution.

When to use
A T-test is appropriate when the population standard deviation is unknown and the sample size is small or moderate.

3. Key Differences
   The Z-test uses the normal distribution, while the T-test uses the t-distribution.
   The Z-test requires known population variance, whereas the T-test does not.

4. Conclusion
   Choosing between a Z-test and a T-test depends on sample size and whether population variance is known.
'''
```

```python
#6.Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

import numpy as np                 # Import NumPy for numerical operations
import matplotlib.pyplot as plt    # Import Matplotlib for plotting

n = 10                             # Number of trials
p = 0.5                            # Probability of success
size = 1000                        # Number of experiments (samples)


data = np.random.binomial(n, p, size) # Generate binomial distribution data


plt.hist(data, bins=range(0, n + 2), edgecolor='black') # Plot histogram of the binomial distribution

plt.xlabel("Number of Successes")  # Label for x-axis
plt.ylabel("Frequency")            # Label for y-axis
plt.title("Binomial Distribution (n=10, p=0.5)")  # Plot title

plt.show()                         # Display the histogram
```

```python
#7. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. 
#Plot the data using Matplotlib.
import numpy as np                      # For numerical operations
import matplotlib.pyplot as plt         # For plotting
from scipy import stats                # For statistical functions

np.random.seed(42)                     # Set seed for reproducibility

# Parameters for the normal distribution
mean = 50                              # True mean
std_dev = 10                           # Standard deviation
sample_size = 100                     # Number of observations


data = np.random.normal(mean, std_dev, sample_size) # Simulate data from a normal distribution


sample_mean = np.mean(data)     # Calculate sample mean and standard error
standard_error = stats.sem(data)


confidence_level = 0.95            # Compute the 95% confidence interval for the mean
confidence_interval = stats.t.interval(
    confidence_level,
    df=sample_size - 1,
    loc=sample_mean,
    scale=standard_error
)


plt.hist(data, bins=15) # Plot the data
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Simulated Normal Data")

plt.show()

# Display results
print("Sample Mean:", sample_mean)
print("95% Confidence Interval:", confidence_interval)
```

```python
#9. Write a Python function to calculate the Z-scores from a dataset and
#visualize the standardized data using a histogram. Explain what the Z-scores represent
#in terms of standard deviations from the mean
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)                 # Calculate mean of the data
    std_dev = np.std(data)               # Calculate standard deviation
    z_scores = (data - mean) / std_dev   # Compute Z-scores
    return z_scores

# Sample dataset
data = np.array([45, 50, 52, 48, 47, 55, 60, 42, 49, 53])

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=8, edgecolor='black')
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Z-scores (Standardized Data)")
plt.show()

# Print Z-scores
print("Z-scores:", z_scores)


```
